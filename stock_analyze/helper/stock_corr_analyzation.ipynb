### stock corr and cov 

```python
# coding: utf-8

import pandas as pd
import numpy as np
from pandas import DataFrame
import pandas_datareader.data as web
import datetime
import MySQLdb

print('imported.')
```

```python
# load stock code and market code from sql
db_conn = MySQLdb.connect(
    host='127.0.0.1',
    port=3306,
    user='admin',
    passwd='sql123456',
    db='stock',
    charset='utf8'
)
sql = 'select name,code,market_code from stock_info where industry_id='\
        '(select id from industry where name="银行")'
stock_info_df = pd.read_sql(sql, con=db_conn)

print(stock_info_df.shape)
stock_info_df.head()
```

```python
start = datetime.datetime(2018,12,24)
# end = datetime.datetime(2018,12,28)
end = datetime.date.today()
all_data = {}
for i in np.arange(0, stock_info_df.shape[0]):
    try:
        stock = stock_info_df.iloc[i]
        request_code = stock['code'] + ('.SS' if int(stock['market_code']) == 1 else '.SZ') 
        all_data[stock['name']] = web.DataReader(request_code,'yahoo',start, end)
    except Exception as e:
#         过会但是未上市的获取时会出错
        print( repr(e) )
    
all_data['常熟银行'].tail()
```

```python
price = DataFrame({tic: data['Adj Close'] for tic, data in all_data.items() })
price.tail()
```

```python
volume = DataFrame({tic: data['Volume'] for tic, data in all_data.items() })
volume.tail()
```

```python
pct_change = price.pct_change()
pct_change.tail()
```

```python
#  两支股票的涨跌幅相关性
pct_change['常熟银行'].corr(pct_change['吴江银行'])
```

```python
pct_change.corrwith(pct_change['常熟银行'])
```

```python
#  两支股票的涨跌幅协方差
pct_change['常熟银行'].cov(pct_change['吴江银行'])
```

```python
# 两两之间的相关性
pct_change.corr()
```

```python
# 两两之间的协方差
pct_change.cov()
```

```python
# 涨跌幅与成交量的关系
pct_change.corrwith(volume)
```

```python
price.corrwith(volume)
```
